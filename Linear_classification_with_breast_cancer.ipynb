### <font color = 'blue'> **5.** Linear classification

After reviewing and understanding main features of the dataset, you will start to implement a linear classification on the dataset. Several steps to follow:

- Train a linear classifier with different optimization algorithms, i.e., gradient descent (GD), Stochastic Gradient descent (SGD) and minibatch gradient descent (GD). Plot the traning loss over epochs for GD, SGD, and minibatch GD, respectively.

- Apply the parameters (or weights **w**) optimized by GD, SGD and mini-batch GD for predicting the test dataset.

- Compute the test accuracy corresponding to the three classifiers trained with GD, SGD, minibatch GD, respectively; and then print the accuracy values.

- Compute the confusion matrix and visualize it as a *heatmap* to evaluate the classification accuracy for test data, see example in https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html.

```python
# Note that each classifier should return the learned parameters and a list of loss values over epochs

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def compute_loss(y_true, y_pred):
    if np.isscalar(y_true):
      n_samples = 1 # sample only one data while using SGD
    else:
      n_samples = len(y_true)
    return -(1 / n_samples) * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

def gradient_w(X, y, w, b):
    N = X.shape[0]
    y_hat = np.dot(X, w) + b
    return 1/N * np.dot(X.T, (y_hat - y))

def gradient_b(X, y, w, b):
    N = X.shape[0]
    y_hat = np.dot(X, w) + b
    return 1/N * np.sum(y_hat - y)
def loss_over_epoch():
    loss_by_epochs = [np.mean(losses[i:i+n_iters]) for i in range(0, len(losses), n_iters)]

# classifier 1: optimized with Gradient Descent

def classifier_GD(X, y, learning_rate, epochs):
    # initialization of parameters (weights: w; bias: b)
    N, D = X.shape
    w = np.zeros(D)
    b = 0
    losses = []
    for i in range(epochs):
        z = np.dot(X, w) + b
        y_prob = 1 / (1 + np.exp(-z)) # sigmoid function after the output
        # compute logistic loss (i.e., cross-entropy loss)
        loss = compute_loss(y, y_prob)
        losses.append(loss) # logistic regression for classification
        # compute gradients
        grad_w = gradient_w(X, y, w, b)
        grad_b = gradient_b(X, y, w, b)
        # update weights and bias
        w -= learning_rate * grad_w
        b -= learning_rate * grad_b
    return w, b, losses


# classifier 2: optimized with Stochastic Gradient Descent

def classifier_SGD(X, y, learning_rate, epochs):
    N, D = X.shape
    w = np.zeros(D)
    b = 0
    losses = []
    for _ in range(epochs):
        epoch_loss = []
        for i in range(N):
            xi, yi = X[i], y[i]
            z = np.dot(xi, w) + b
            y_prob = 1 / (1 + np.exp(-z)) # sigmoid function after the output
            # compute logistic loss (i.e., cross-entropy loss)
            loss = compute_loss(yi, y_prob)
            epoch_loss.append(loss) # logistic regression for classification
            # compute gradients
            grad_w = gradient_w(xi.T, yi, w, b)
            grad_b = gradient_b(xi.T, yi, w, b)
            # update weights and bias
            w -= learning_rate * grad_w
            b -= learning_rate * grad_b
        losses.append(np.mean(epoch_loss))
    return w, b, losses


# classifier 3: optimized with Mini-batch Gradient Descent

def classifier_mGD(X, y, learning_rate, epochs, batch_size):
    N, D = X.shape
    w = np.zeros(D)
    b = 0
    losses = []
    for _ in range(epochs):
        idx = np.random.permutation(N)
        epoch_loss = []
        for i in range(0, N, batch_size):
            batch_indices = idx[i:i+batch_size]
            X_batch, y_batch = X[batch_indices], y[batch_indices]
            z = np.dot(X_batch, w) + b
            y_prob = 1 / (1 + np.exp(-z)) # sigmoid function after the output
            # compute logistic loss (i.e., cross-entropy loss)
            loss = compute_loss(y_batch, y_prob)
            epoch_loss.append(loss) # logistic regression for classification
            # compute gradients
            grad_w = gradient_w(X_batch, y_batch, w, b)
            grad_b = gradient_b(X_batch, y_batch, w, b)
            # update weights and bias
            w -= learning_rate * grad_w
            b -= learning_rate * grad_b
        losses.append(np.mean(epoch_loss))
    return w, b, losses

# Prediction function
#  (input: data, parameters; return: predicted labels for test data)
def predict(X, w, b):
    z = np.dot(X, w) + b
    predictions = 1 / (1 + np.exp(-z))
    y_pred = [1 if i > 0.5 else 0 for i in predictions]
    return np.array(y_pred)


# Training  (call three different classifers here)
#### epochs = 1000, you may want to test two different batch sizes (16, 32);
#### and differnet learning rates (0.001, 0.01) to find the best parameters to
#### achieve good performance.

learning_rate = [0.001, 0.01]
epochs = 1000
batchsize = [16, 32]

gd_pred = []
sgd_pred = []
mgd_pred = []
N, D = X_train.shape
for lr in learning_rate:
  w_gd, b_gd, gd_losses = classifier_GD(X_train, y_train, lr, epochs)
  w_sgd, b_gd, sgd_losses = classifier_SGD(X_train, y_train, lr, epochs)

  # Plotting loss curves over epochs for GD, SGD, minibatch GD algorithms
  plt.figure(figsize = [10,5])
  plt.plot(gd_losses, label='Gradient Descent (minimum loss: %.5f)'%gd_losses[-1])
  plt.legend()
  plt.xlabel('Epochs', size = 14)
  plt.ylabel('Loss', size = 14)
  plt.title('learning rate: %s'%str(lr))

  plt.figure(figsize = [10,5])
  plt.plot(sgd_losses, label='Stochastic Gradient Descent (minimum loss: %.5f)'%sgd_losses[-1])
  plt.legend()
  plt.xlabel('Epochs', size = 14)
  plt.ylabel('Loss', size = 14)
  plt.title('learning rate: %s'%str(lr))

  # Test the model (call the prediction function)
  y_pred_gd = predict(X_test, w_gd, b_gd)
  y_pred_sgd = predict(X_test, w_sgd, b_gd)
  gd_pred.append(y_pred_gd)
  sgd_pred.append(y_pred_sgd)

  # Calculate mean test accuracy
  accuracy_gd = np.mean(y_pred_gd == y_test)
  accuracy_sgd = np.mean(y_pred_sgd == y_test)
  print(f"Accuracy using Gradient Descent (lr = {str(lr)}): {accuracy_gd*100:.2f}%")
  print(f"Accuracy using Stochastic Gradient Descent (lr = {str(lr)}): {accuracy_sgd*100:.2f}%")

  # Try different batch sizes
  for b in batchsize:
    w_mgd, b_mgd, mgd_losses = classifier_mGD(X_train, y_train, lr, epochs, batch_size = b)

    # Plotting loss curves over epochs for GD, SGD, minibatch GD algorithms
    plt.figure(figsize = [10,5])
    plt.plot(mgd_losses, label='Mini-batch GD (batchsize = %d, minimum loss: %.5f)'%(b, mgd_losses[-1]))
    plt.legend()
    plt.xlabel('Epochs', size = 14)
    plt.ylabel('Loss', size = 14)
    plt.title('learning rate: %s'%str(lr))

    # Test the model (call the prediction function)
    y_pred_mgd = predict(X_test, w_mgd, b_mgd)
    mgd_pred.append(y_pred_mgd)

    # Calculate mean test accuracy
    accuracy_mgd = np.mean(y_pred_mgd == y_test)
    print(f"Accuracy using minibatch GD with learning rate {str(lr)} and batch size {b}: {accuracy_mgd*100:.2f}%")
```
